```python
credit_card_distribution = credit_card_data['Class'].value_counts()

# Create a pie chart
labels = ['Legit Transaction', 'Fraudulent Transaction']
sizes = credit_card_distribution.values  # Corresponding counts
colors = ['blue', 'red']  # Custom colors
explode = (0.1, 0)  # Explode the first slice (if Male is first)

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90)

# Equal aspect ratio ensures the pie chart is circular
plt.axis('equal')
plt.title('Credit card Distribution')
plt.show()
```

```python
# Seperating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Under-Sampling - Building a sample dataset containing similar distribution of normal transactions and Fradulent transactions
print(fraud.shape)
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Combining new dataset of 492 normal and 492 fraudulent data
new_dataset = pd.concat([legit_sample, fraud], axis=0)
# the axis=0 parameter specifies that the two dataframes (legit_sample and fraud) should be concatenated vertically (i.e., one on top of the other)
```

```python
new_dataset.groupby('Class').mean()
# This shows that nature of dataset has not changed as values are similar to what we had in the orignal dataset
```

```python
# Splitting dataset into  features and class
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
# Splitting dataset into training and testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# Training model
model.fit(X_train, Y_train)
```

```python
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, Y_train)
print('Accuracy on training data: ', training_data_accuracy)
```

```python
testing_data_prediciton = model.predict(X_test)
testing_data_prediciton = accuracy_score(testing_data_prediciton, Y_test)
print('Accuracy on testing data: ', testing_data_prediciton)
```
